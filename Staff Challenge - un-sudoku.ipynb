```ruby
require 'sciruby'
require 'set'
```

```ruby
def col(arr, n)
    Set.new arr.transpose[n]
end

def row(arr, n)
    Set.new arr[n]
end

def cell_block(matrix, i, j)
    start_i = i / 3 * 3
    start_j = j / 3 * 3
    values = []
    ii, jj = start_i, start_j
    
    3.times do
       3.times do
           values << matrix[ii][jj]
           jj += 1
       end
        ii += 1
        jj = start_j
    end
    values.to_set
end

def count_zeros(matrix)
    zeros = 0
    
    matrix.each.with_index do |row, i|
        row.each.with_index do |value, j|
            if value == 0
                zeros += 1
            end
        end
    end
    zeros
end
```

```ruby
def can_be?(matrix, i, j, candidate)
    !((row(matrix, i) | col(matrix, j)).include? candidate)
end

def available_coords_in_cell_block(matrix, i, j)
    start_i = i / 3 * 3
    start_j = j / 3 * 3
    ii, jj = start_i, start_j
    values = []
    3.times do
       3.times do
           values << [ii, jj] if matrix[ii][jj] == 0
           jj += 1
       end
        ii += 1
        jj = start_j
    end
    values
end

def relaxed_can_be_removed(matrix, i, j)
    possibilities = Set.new([*(1..9)])
    value = matrix[i][j]
    row = row(matrix, i) - [value]
    col = row(matrix, j) - [value]
    cb = cell_block(matrix, i, j) - [value]
    (possibilities - row - col - cb).to_a == [value]
end

def this_too_can_be_removed(puzzle, i, j)
    candidate = puzzle[i][j]
    empty_spots = available_coords_in_cell_block(puzzle, i, j)
    empty_spots.map do |i, j|
        !can_be?(puzzle, i, j, candidate)
    end.all?
end
```

```ruby
def removable? matrix, i, j
    can_be_removed(matrix, i, j) || relaxed_can_be_removed(matrix, i, j)
end
```

```ruby
def remove(matrix)
    removed = false
    
    matrix.each.with_index do |row, i|
        row.each.with_index do |value, j|
            next if value == 0
            
            if removable?(matrix,i, j) && !removed
                removed = true
                new_value = 0
            else
                new_value = value
            end
            matrix[i][j] = new_value
        end
    end
    return removed
end
```

```ruby
def reduce(puzzle)
    keep_removing = true
    while(keep_removing)
        keep_removing = remove(puzzle)
    end
end

reduce(puzzle)
```
